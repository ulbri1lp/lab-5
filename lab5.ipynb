```python
#import modules
import requests
import json
```

Question 1:

```python
#base url
url = 'https://www.loc.gov'

#combining base url and list of resources
uri1 = url + '/resource/cph.3f05183/'  # item 1
uri2 = url + '/resource/fsa.8d24709/'  # item 2
uri3 = url + '/resource/highsm.64003/' # item 3

#print answer
print(uri1, uri2, uri3)
```

Question 2

```python
#define parameters
parameters = {'fo': 'json'}

#request from item 1
r1 = requests.get(uri1, params=parameters)

#display returned data
r1.json()
```

```python
#write each item's json data to their own file
item1 = json.dumps(r1.json())
with open("item1.json", "w") as f:
    f.write(item1)

item2 = json.dumps(r2.json())
with open("item2.json", "w") as f:
    f.write(item2)

item3 = json.dumps(r3.json())
with open("item3.json", "w") as f:
    f.write(item3)
```
