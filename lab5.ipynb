Question 4

```python
#prompt user to search
print('What would you like to search?')

#user input
search_term = input()

#define url addition
search_endpoint = '/search'

#define parameters
parameters = {'fo':'json', 'q': search_term}

#request from search
r4 = requests.get(url + search_endpoint, params = parameters)

#display returned data
r4_json = r4.json()

with open(search_term + '_result.json', 'w', encoding='utf-8') as f:
    json.dump(r4_json, f, indent=2)
```
